```python
#W5A3
def so_nguyen_to(a: int):
    if a <= 3 and a > 1:
        return True
    else:
        for i in range (2, int(a**(1/2) + 1)):
            if a%i == 0:
                return False
                break
        return True
print(so_nguyen_to(11))
```

```python
#w5A4
def so_hoan_hao(a: int):
    b = 1
    for i in range (2,a):
        if a % i == 0:
            b = b + i
    if b == a:
        return True
    else:
        return False
print(so_hoan_hao(6))
```

```python
#W5A5
def tim_so(A: list[int], k: int):
    for i in range (len(A)):
        if A[i] == k:
            return i + 1
    return -1
D = [1, 2, 3, 5, 7]
b = 3
ket_qua = tim_so(D, b)
print(ket_qua)
```

```python
#W5A6
def giai_thua(n: int):
    x = 1
    for i in range (1, n + 1):
        x = x*i
    return x
a = 4
ket_qua = giai_thua(a)
print(ket_qua)
        
```

```python
#W5A8
def khoang_cach_hamming(n: int, m: int): 
    nhi_phan_1 = ''
    nhi_phan_2 = ''
    a = 0
    b = 0
    c = 0
    if n == 0:
        nhi_phan_1 = '0'
    else:
        while n > 0:
            a = n%2
            nhi_phan_1 = str(a) + nhi_phan_1
            n = n//2
    if m == 0:
        nhi_phan_2 = '0'
    else:
        while m > 0:
            b = m%2
            nhi_phan_2 = str(b) + nhi_phan_2
            m = m//2
    max_len = max(len(nhi_phan_1), len(nhi_phan_2))
    nhi_phan_1 = nhi_phan_1.zfill(max_len)
    nhi_phan_2 = nhi_phan_2.zfill(max_len) # .zfill(độ dài) thêm số 0 vào trước để đạt 'độ dài' yêu cầu
    for i in range(0,max_len):
        if nhi_phan_1[i] != nhi_phan_2[i]:
            c = c + 1
    return c
n = 1
m = 4
print(khoang_cach_hamming(n, m))
```
